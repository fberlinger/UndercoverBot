# Experiment: Hop Count

A seed starts a hop count and gets back the number of hops across the network. We also measure the number of messages it takes to complete the count.

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import matplotlib

matplotlib.rcParams['figure.figsize'] = [12, 8]

import math
import numpy as np

from interaction import Interaction
from environment import Environment
from fish import Fish
from channel import Channel
from observer import Observer

from utils import generate_distortion, generate_fish, init_simulation, run_simulation
```

### Center, static
The seed robot which is marked by a star at the center of the network starts a hop count. All robots keep their positions.

```python
"""Hop count: center, static
"""

from events import StartHopCount

num_fish = 40
clock_freq = 5  # in Hz
single_time = 2 * num_fish + 0 # run time of one trial
offset_time = 1  # offset in sec when the first trial starts
num_trials = 1
final_buffer = 10
run_time = single_time * num_trials + final_buffer # in seconds
run_time /= (clock_freq * 0.9)
size_dist = 100
center = size_dist / 2.0
fish_pos = np.zeros((num_fish, 2))
spread = 10
offset = 0
fish_pos = spread * np.random.rand(num_fish, 2) + center - spread/2.0
fish_pos[0] = np.array([center + offset, center + offset])

channel, environment, fish, interaction, observer = init_simulation(
    clock_freq=clock_freq,
    single_time=single_time,
    offset_time=offset_time,
    num_trials=num_trials,
    final_buffer=final_buffer,
    run_time=run_time,
    num_fish=num_fish,
    size_dist=size_dist,
    center=center,
    spread=spread,
    fish_pos=fish_pos,
    lim_neighbors=[1, math.inf],
    neighbor_weights=1.0,
    fish_max_speeds=1,
    noise_magnitude=0.05,
    conn_thres=3.5,
    prob_type='binary',
    dist_type='none',
    verbose=False
)

for i in range(num_trials):
    observer.instruct(
        event=StartHopCount(), 
        rel_clock=i * single_time + offset_time,
        fish_id=0
    )
observer.activate_reset()
observer.study('hop_count')

run_simulation(fish=fish, observer=observer, run_time=run_time, dark=True, white_axis=False, no_legend=True)
```
